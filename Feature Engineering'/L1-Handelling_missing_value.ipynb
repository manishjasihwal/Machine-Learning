## Technique use to handle Missing Values

# 1-Mean Value Imputation

```python
sns.distplot(df['age'])
```

```python
df.age.isnull().sum()
```

```python
df['Age_Mean']=df['age'].fillna(df['age'].mean())
```

```python
df[['Age_Mean' , 'age']]
```

## This technique works well when your data is normally distributed

## 2-Median value imputation

```python
df['Age_Median']=df['age'].fillna(df['age'].median())
```

```python
df[['Age_Median' , 'Age_Mean' ,  'age']]
```

## If you have outlier in datasets use this technique.

### 3-Mode Value Imputation -- Categorical

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['age'].notna()]['embarked'].mode()
```

```python
mode=df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode
```

```python
df['embarked_mode']=df['embarked'].fillna(mode)
```

```python
df[['embarked_mode' , 'embarked']]
```

```python
df['embarked_mode'].isnull().sum()
```
