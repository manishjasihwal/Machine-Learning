```python
from sklearn.datasets import make_blobs
import numpy as np
```

```python
#Creating 400 dsamples and creating 2 different feature like x1 and x2 with one center
X , _=make_blobs(n_samples=400 , n_features=2 , centers=1 , shuffle=True , random_state=2)
```

```python
import seaborn as sns
sns.scatterplot(x=X[:,0] , y=X[:,1])
```

```python
X
```

Plotting X to histogram .  [-1.82398534e+00, -9.29717362e+00] This means there is two column  c0 and c1. plotting seperately histogram for column-0 and column-1

```python
import matplotlib.pyplot as plt
fig , ax = plt.subplots(nrows=1 , ncols=2 , figsize=(11 , 3))
sns.displot(X[:,0] , ax=ax[0] , color="red")
sns.displot(X[:,1] , ax=ax[1] , color="blue")
```

From Above plot , we can observe that plot is gaussain plot

```python
from sklearn.covariance import EllipticEnvelope
```

```python
# Create EllipticEnvelope
elpenv = EllipticEnvelope(contamination=0.025 , 
                            random_state=1)
```

Now on my dataset X , go and execute EllipticEnvelope

```python
pred=elpenv.fit_predict(X)
```

```python
pred
```

-1 are outlier

```python
# This are the indexes , where values was showing to -1 
index = np.where(pred == -1)
```

```python
# I want to check actual value
values=X[index]
```

```python
values
```

Now , I want to plot in scatter plot

```python
sns.scatterplot(x= values[:,0] ,
                y= values[:,1] , 
                color='r' 
                )
```
